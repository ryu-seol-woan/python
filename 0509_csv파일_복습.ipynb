```python
#함수화 하기

def opencsv(filename):
  f = open(filename, 'r')
  reader = csv.reader(f)
  output = []
  for i in reader:
    output.append(i)
  return output
```

```python
# 만약, 문자와 숫자가 섞여있다면?
# 데이터가 밀리는 경우는 부지기수.

p = ['123강남구', '456종로구', '11,093', '27,394']
k = []
for j in p:
  if re.search('[a-z가-힣]', j): #j에서 소문자 알파벳 또는 한글 문자가 있다면
    k.append(j)
  else:
    k.append(float(re.sub(',' , '', j))) 
```

```python
def switch(listname):
  for i in listname:
    for j in i:
      try:
        i[i.index(j)] = float(re.sub(',','',j))
      except:
        pass
  return listname
```
