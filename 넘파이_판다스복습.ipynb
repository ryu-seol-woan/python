# 난수 생성 함수
random

```python
import numpy as np
```

```python
a = np.random.rand(2,2) # 0 ~ 1 사이에 값 랜덤
print(a)
```

```python
b = np.random.randn(2,2) # 평균이 0, 분산이 1: 표준 정규 분포
print(b)
```

```python
c = np.random.randint(0,10, size=5) # 랜덤한 정수값 생성
print(c)
```

```python
c2 = np.random.randint(0,10, size=(2,2))
print(c2)
```

```python
d = np.arange(5)
np.random.shuffle(d) #복사본 새성이 x 원본값 변함
print(d)
```

```python
# choice = np.random.choice(d,5) #중복을 허용
# choice = np.random.choice(d,5,replace=False) #중복 허용 x
choice = np.random.choice(d, 5, p = [0,0.4,0.2, 0.3, 0.1])
#p값을 100% 로 설정
print(choice)
```

```python
#permutation 순열 -> 수정해서 파일 전함
a = np.arange(10)
a_p = np.random.permutation(a)
print(a_p)
print(a)
```

```python
#정규분포를 따르는 난수 생성
# a = np.random.normal('평균','표준편차','갯수')
a = np.random.normal(0,1,3)
print(a)
```

# Series 시리즈

```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4]) # 리스트
print(s)
```

# series 산술연

```python
print(series + 100)
print(series - 100)
print(series * 100)
print(series ** 100)
print(series / 100)
print(series % 100)
print(series // 100)
# 곱하기: *, 제곱: **, 나누기(float 몫): /, 나머지: %, 나누기(int 몫): //
```

```python
s = pd.Series([10,20,30,40], index=['a','b','c','d'])
```

```python
print(s)
```

```python
print(s.loc['b']) # loc: 행에 접근 -> 레이블(특정 인덱스)에 접근하기 위해서 -> 슬라이싱 범위 다포
print(s.loc['b':'d'])
```

```python
print(s.iloc[1]) # iloc : 행에기반한 열에 접근
```

```python
s.iloc[-1]
```

```python
print(s.at['a']) # at loc 비슷, 하나의 값만 가져올 때
```

```python
print(s.iat[0]) # at 에 기반
```

```python
print(s.get('e', default=-1)) # 없을 때 값 처리 방법이 다름, error가 나지 않고 없다는 걸 알려줌
```

# 결측값(NaN, None)
