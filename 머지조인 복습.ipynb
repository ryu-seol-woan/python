```python
df1 = pd.DataFrame([['A0', 'A1', 'A2', 'A3'],
                     ['B0', 'B1', 'B2', 'B3'],
                     ['C0', 'C1', 'C2', 'C3'],
                     ['D0', 'D1', 'D2', 'D3']],
                    columns=list('ABCD'))
 
df2 = pd.DataFrame([['A4', 'A5', 'A6', 'A7'],
                     ['B4', 'B5', 'B6', 'B7'],
                     ['C4', 'C5', 'C6', 'C7'],
                     ['D4', 'D5', 'D6', 'D7']],
                   columns=['A', 'B', 'C', 'D'],
                   index=[4, 5, 6, 7])

df3 = pd.DataFrame([['A8', 'A9', 'A10', 'A11'],
                     ['B8', 'B9', 'B10', 'B11'],
                     ['C8', 'C9', 'C10', 'C11'],
                     ['D8', 'D9', 'D10', 'D11']],
                    columns=list('ABCD'),
                    index=[8, 9, 10, 11])
```

```python
print(df1)
print(df2)
print(df3)
```

```python
pd.concat([df1,df2,df3],axis = 1)
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                  'F': ['F2', 'F3', 'F6', 'F7']}, index=[2, 3, 6, 7])
```

```python
pd.concat([df1,df4], axis=1, join = 'outer')
```

```python
pd.concat([df1, df4], axis=1, join='inner') # inner: 교집합
```

```python
pd.concat([df1,df4], axis=1).reindex(df1.index) # left join
```

```python
pd.concat([df1,df4], axis=1).reindex(df4.index) # left join
```
